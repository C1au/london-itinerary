```python
import folium

# Definire i colori per ogni giorno
day_colors = {
    "1": "red",
    "2": "blue",
    "3": "green",
    "4": "purple",
    "5": "orange"
}

# Creare una mappa centrata su Londra
london_coords = (51.5074, -0.1278)
mymap = folium.Map(location=london_coords, zoom_start=12)

# Dati per i punti di interesse
places = {
    "Day 1": [
        {"name": "Mercato centrale Mayfair", "coords": (51.5127, -0.1430)},
        {"name": "Piccadilly Circus", "coords": (51.5101, -0.1347)},
        {"name": "National Gallery", "coords": (51.5089, -0.1283)},
        {"name": "Baker Street", "coords": (51.5235, -0.1586)},
        {"name": "Madame Tussauds", "coords": (51.5230, -0.1544)}
    ],
    "Day 2": [
        {"name": "Victoria and Albert Museum", "coords": (51.4966, -0.1722)},
        {"name": "Museo di Storia Naturale e il Museo della Scienza", "coords": (51.4967, -0.1764)},
        {"name": "Holland Park", "coords": (51.5048, -0.2015)},
        {"name": "Regent’s Park", "coords": (51.5313, -0.1569)}
    ],
    "Day 3": [
        {"name": "Queen’s Gallery di Buckingham Palace", "coords": (51.4992, -0.1425)},
        {"name": "Abbazia di Westminster", "coords": (51.4993, -0.1273)},
        {"name": "Big Ben", "coords": (51.5007, -0.1246)},
        {"name": "Chiesa di Saint Margaret", "coords": (51.4998, -0.1281)},
        {"name": "Tower Bridge", "coords": (51.5055, -0.0754)}
    ],
    "Day 4": [
        {"name": "Castello di Windsor", "coords": (51.4839, -0.6044)},
        {"name": "Kew Gardens", "coords": (51.4787, -0.2950)},
        {"name": "Richmond Park", "coords": (51.4427, -0.2733)}
    ],
    "Day 5": [
        {"name": "British Museum", "coords": (51.5194, -0.1270)},
        {"name": "La Torre di Londra", "coords": (51.5081, -0.0759)},
        {"name": "Mercato di Portobello Road", "coords": (51.5175, -0.2066)}
    ]
}

# Aggiungere i marker alla mappa
for day, locations in places.items():
    for place in locations:
        folium.Marker(
            location=place["coords"],
            popup=place["name"],
            icon=folium.Icon(color=day_colors[day.split()[1]])
        ).add_to(mymap)

# Salvare la mappa come file HTML
mymap.save("London_Itinerary_Map.html")

# Visualizzare la mappa direttamente nel notebook
from IPython.display import IFrame
IFrame('London_Itinerary_Map.html', width=700, height=500)
```
